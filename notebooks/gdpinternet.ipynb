```python
# Q24. Stretch Question: Use the qcut function from pandas (https://pandas.pydata.org/docs/reference/api/pandas.qcut.html) to divide countries in gdp_per_capita_2014 into three groups based on their GDP per capita values.
# Label these groups as "Low", "Medium", and "High". Put these labels in a new column, named "GDP_group".
gdp_and_internet_use_2014.loc[:, "GDP_group"] = pd.qcut(
    gdp_and_internet_use_2014["GDP_Per_Capita"], 
    q=3,  
    labels=["Low", "Medium", "High"]
)
```

```python
gdp_and_internet_use_2014[["Country", "GDP_Per_Capita", "GDP_group"]].head())
```

```python
# Q25. Stretch Question: How does the median internet users percentage compare for the three gdp groups? - wealthier countries tend to have higher internet
# percentage. gap is large between low and medium GDP groups. 

median_internet_users = gdp_and_internet_use_2014.groupby("GDP_group", observed=False)["Internet_Users_Pct"].median()
median_internet_users

```
